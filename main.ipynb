```python
from matplotlib import pyplot as plt
import numpy as np

def random_walk_symulation_1d(N: int, p: float, M: int):
    
    trajectories=[]
    
    for i in range(M):
        steps = np.random.choice([1,-1], size=N, p=[p,1-p])
        trajectory = np.cumsum(np.insert(steps, 0, 0))
        trajectories.append(trajectory)
    
    all_positions = np.array(trajectories)
    avg_positions = np.mean(all_positions, axis=0)
    final_positions = [trajectory[-1] for trajectory in trajectories]
    avg_final_position = avg_positions[-1]
    
    plt.figure(figsize=(10,6))
    for i, trajectory in enumerate(trajectories):
        plt.plot(trajectory, label=f'Trajectory {i+1}')
        plt.scatter(len(trajectory) - 1, trajectory[-1], color=plt.gca().lines[-1].get_color(), s=50, label=f'End {i+1}')
    
    plt.plot(avg_positions, 'k--', linewidth=2, label='Average Position')
    plt.title(f"Random Walks (N={N}, p={p}, M={M})")
    plt.xlabel("Steps")
    plt.ylabel("Position")
    plt.legend()
    plt.grid()

    print("Final positions of each trajectory: ", final_positions)
    print("Final postion of avg trajectory: ", avg_final_position)
    return avg_positions, trajectories
```

```python
N = 100  
p = 0.025  
M = 10    

avg_positions, trajectories = random_walk_symulation_1d(N, p, M)
```

```python
confidence_level = [0.1, 0.05, 0.025]
avg_positions = [-80.2, -89.4, -94.6]

plt.scatter(confidence_level, avg_positions, marker = 'o', label='Średnia wartość pozycji końcowych')

for p, mean in zip(confidence_level, avg_positions):
    plt.text(p, mean, f"({p}, {mean})", fontsize=10, ha='left')
    

plt.title("Średnia wartość pozycji dla 100 kroków w zależności od przedziału ufności")
plt.xlabel("Przedział ufności (p)")
plt.ylabel("Średnia wartość pozycji")
plt.grid(0)
plt.legend()
```

```python
positions_p_025 = [-92, -92, -98, -100, -96, -100, -88, -96, -94, -90]
positions_p_05 = [-94, -82, -82, -98, -84, -94, -80, -90, -96, -94]
positions_p_1 = [-80, -86, -78, -82, -82, -76, -80, -86, -68, -84]

fig, axes = plt.subplots(1, 3, figsize=(15, 5))

def plot_histogram(ax, data, bins, color, label, title):
    ax.hist(data, bins=bins, color=color, alpha=0.7, edgecolor='black', label=label)
    ax.set_title(title)
    ax.set_xlabel("Pozycje końcowe dla N=100 kroków")
    ax.set_ylabel("Ilość wystąpień")
    ax.legend()

plot_histogram(axes[0], positions_p_025, bins=10, color='blue', label='p=0.025', title="Histogram dla p=0.025")
plot_histogram(axes[1], positions_p_05, bins=10, color='green', label='p=0.05', title="Histogram dla p=0.05")
plot_histogram(axes[2], positions_p_1, bins=10, color='red', label='p=0.1', title="Histogram dla p=0.1")

plt.tight_layout()

```
